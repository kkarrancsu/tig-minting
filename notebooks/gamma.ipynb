```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gamma_exponential(n: int, a, b, c: float, n_offset: int = 0, y_offset: float = 0) -> float:
    # if n < n_offset:
    #     # linearly increase from 0 to y_offset
    #     return y_offset + (n-n_offset) * (y_offset - 0) / (n_offset - 0)
    return a*(( 1-b*np.exp(-c * (n - n_offset)) ) * (1-y_offset) + y_offset)
```

```python
n_values = np.arange(0, 250)  # Show up to 500 challenges
a = 1.10
b = 0.85
c = 0.02333
gamma_start_n = 0
gamma_start_y = 0
gamma_func_params = (a, b, c, gamma_start_n, gamma_start_y)

gamma_values = []
for nn in n_values:
    gamma_values.append(gamma_exponential(nn, *gamma_func_params))
```

```python
plt.figure(figsize=(7, 4))
plt.plot(n_values, gamma_values, label='Gamma Function')
plt.axvline(x=100, color='r', linestyle='--')
plt.axhline(y=1, color='r', linestyle='--', label='Gamma = 1')
plt.axhline(y=0.5, color='g', linestyle='--')
plt.axvline(x=20, color='g', linestyle='--')

plt.xlabel('Number of Challenges')
plt.ylabel('Gamma Value')
plt.title('Gamma Function for Exponential Decay')
plt.legend()
plt.show()
```
